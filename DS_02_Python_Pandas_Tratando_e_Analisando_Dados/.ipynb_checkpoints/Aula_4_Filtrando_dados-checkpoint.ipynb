# Relatório III

   ## Imovéis residenciais

```python
import pandas as pd
```

```python
dados = pd.read_csv("aluguel.csv", sep=';')
                
dados.head(10)
```

```python
list(dados['Tipo'].drop_duplicates())
```

```python
residencial = ['Quitinete',
 'Casa','Apartamento',
 'Casa de Condomínio', 'Casa de Vila']
```

```python
residencial
```

```python
selecao = dados['Tipo'].isin(residencial)
selecao
```

```python
dados_resid = dados[selecao] #criando um novo DataFrane com a selecao
#passo o DF a info de que eu só quero os valores que são TRUE para o Tipo residencial
dados_resid
```

```python
dados_resid.info()
```

```python
dados_resid.describe()
```

```python
list(dados_resid['Tipo'].drop_duplicates())
```

```python
dados_resid.shape
```

```python
dados_resid.shape[0]
```

```python
dados_resid.index = range(dados_resid.shape[0])
```
